```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("https://github.com/ezioauditore-tech/AI/raw/main/datasets/Mall_Customers.csv")
```

```python
df
```

```python
features=['Annual Income (k$)','Spending Score (1-100)']
x=df[features]
```

```python
x_std=StandardScaler().fit_transform(x)
```

```python
dbscan=DBSCAN(eps=0.5,min_samples=5)
clusters=dbscan.fit_predict(x_std)
```

```python
centroids=pd.DataFrame(x_std).groupby(clusters).mean().values
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'],c=clusters,cmap='viridis',edgecolors='k',s=50)
plt.title('DBSCAN clustering of mall customer data')
plt.ylabel('Spending Score (1-100)')
plt.xlabel('Annual Income (k$')
```
