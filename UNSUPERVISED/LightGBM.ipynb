```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Breast_cancer_data.csv")
```

```python
y=df['diagnosis']
```

```python
df.drop(columns='diagnosis',axis=1,inplace=True)
```

```python
x=df
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
  import lightgbm as lgb
  clf=lgb.LGBMClassifier(boosting_type='goss',learning_rate=0.05,num_leaves=31,n_estimators=100,)
  clf.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
predictions = clf.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy*100)
```

```python
model_light=lgb.LGBMClassifier()
from sklearn.model_selection import GridSearchCV
param_grid={
    'learning_rate':[0.01,0.05,0.2],
    'num_leaves':[31,30,25],
    'boosting_type':['goss','gbdt','dart'],
    'reg_lambda':[0.1,0.5,0.9],
    'reg_alpha':[0.1,0.5,0.9],
    'n_estimators':[100,200,300]
}
grid_search=GridSearchCV(estimator=model_light,param_grid=param_grid,scoring='accuracy',cv=6,verbose=1)
```

```python
grid_search.fit(x_train,y_train)

print("best parameters: ",grid_search.best_params_)
print("best accuracy: ",grid_search.best_score_)
```
