```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv("/content/Titanic-Dataset.csv")
data
```

```python
data.isnull().sum()
```

```python
data['Cabin'].value_counts()
```

```python
data=pd.DataFrame(data)
```

```python
data['Cabin'].replace(np.nan,'B96 B98',inplace=True)
data['Cabin']
```

```python
data['Age'].replace(np.nan,data['Age'].mean(),inplace=True)
```

```python
data['Embarked'].replace(np.nan,'S',inplace=True)
```

```python
dataf=data[['Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked','Cabin']]
dataf.head()
```

```python
dataf["Sex"]=dataf["Sex"].replace(to_replace=dataf["Sex"].unique(),value=[0,1])
dataf
```

```python
dataf=pd.get_dummies(dataf,drop_first=True)
dataf
```

```python
y=dataf["Survived"]
x=dataf[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked_Q','Embarked_S','Cabin_A14','Cabin_A16','Cabin_A19','Cabin_A20','Cabin_A23','Cabin_A24','Cabin_A26','Cabin_A31','Cabin_A32','Cabin_A34','Cabin_A36']]
```

```python
from sklearn import linear_model
```

```python
model=linear_model.LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
print("Total ",len(x))
print("Train size",len(x_train))
print("Test size",len(x_test))
```

```python
model.fit(x_train,y_train)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
y_predicted=model.predict(x_test)
y_predicted
```

```python
print("test score",model.score(x_test,y_test)*100)
print("train score",model.score(x_train,y_train)*100)
```
