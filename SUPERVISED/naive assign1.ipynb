```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("/content/Titanic-Dataset.csv")
data
```

```python
data.isnull().sum()
```

```python
data['Cabin'].value_counts()
```

```python
data=pd.DataFrame(data)
```

```python
data['Cabin'].replace(np.nan,'B96 B98',inplace=True)
data['Cabin']
```

```python
data['Cabin'].isnull().sum()
```

```python
data['Age'].replace(np.nan,data['Age'].mean(),inplace=True)
```

```python
data['Age'].isnull().sum()
```

```python
data['Embarked'].value_counts()
```

```python
data['Embarked'].replace(np.nan,'S',inplace=True)
```

```python
data['Embarked'].isnull().sum()
```

```python
dataf=data[['Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked','Cabin']]
dataf.head()
```

```python
dataf["Sex"]=dataf["Sex"].replace(to_replace=dataf["Sex"].unique(),value=[0,1])
dataf
```

```python
dataf=pd.get_dummies(dataf,drop_first=True)
dataf
```

```python
train_y=dataf["Survived"]
train_x=dataf[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked_Q','Embarked_S','Cabin_A14','Cabin_A16','Cabin_A19','Cabin_A20','Cabin_A23','Cabin_A24','Cabin_A26','Cabin_A31','Cabin_A32','Cabin_A34','Cabin_A36']]
```

```python
from sklearn.model_selection import train_test_split
train_data,val_data,train_target,val_target=train_test_split(train_x,train_y,train_size=0.8)
```

```python
model=GaussianNB()
model.fit(train_data,train_target)
val_pred=model.predict(val_data)
```

```python
from sklearn.metrics import accuracy_score
print("Model Accuracy:{0:0.4f}".format(accuracy_score(val_target,val_pred)*100)+"%")
```
